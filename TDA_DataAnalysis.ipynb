```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('kindle_data-v2.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
print("Missing values:\n", df.isnull().sum())

```

```python
df.hist(figsize=(6, 5))
plt.suptitle('Histograms of Numerical Columns', y=0.95)
plt.show()
```

```python
categorical_columns = ['isKindleUnlimited', 'isBestSeller', 'isEditorsPick', 'isGoodReadsChoice']
for column in categorical_columns:
    sns.countplot(x=column, data=df)
    plt.title(f'Countplot for {column}')
    plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='category_name', y='price', data=df)
plt.title('Boxplot of Price by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='category_name', y='price', data=df)
plt.title('Average Price by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x='isBestSeller', hue='isKindleUnlimited', data=df)
plt.title('Kindle Unlimited vs. Bestsellers')
plt.show()
```

```python
top_authors = df[df['isBestSeller'] == True]['author'].value_counts().head(10)
plt.figure(figsize=(6, 3))
top_authors.plot(kind='bar')
plt.title('Top Authors of Bestsellers')
plt.xlabel('Author')
plt.ylabel('Number of Bestsellers')
plt.show()
```
